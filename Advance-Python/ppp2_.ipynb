# inheritance

```python
class PARENT:  # single inheritance
  def shop(self):
    print('shop')
class SON(PARENT):
  def job(self):
    print('job')
obj=SON()
obj.job()
obj.shop()
```

```python
class GRANDFATHER:  # multi level inheritance
  def shop(self):
    print('shop')
class FATHER(GRANDFATHER):
  def land(self):
    print('land')
class SON(FATHER):
  def job(self):
    print('job')
obj=SON()
obj.job()
obj.land()
obj.shop()
```

```python
class FATHER:  # multiple inheritance
  def shop(self):
    print('shop')
class MOTHER:
  def land(self):
    print('land')
class SON(FATHER,MOTHER):
  def job(self):
    print('job')
obj=SON()
obj.job()
obj.land()
obj.shop()
```

```python
class FATHER:  # hierarichal inheritance
  def shop(self):
    print('shop')
class B1(FATHER):
  def job(self):
    print('job')
class B2(FATHER):
  def land(self):
    print('land')
class B3(FATHER):
  def bike(self):
    print('bikr')
obj=B3()
obj.bike()
obj.shop()

obj2=B2()
obj2.land()
obj2.shop()

obj3=B1()
obj3.job()
obj3.shop()
```

```python
class A:  # hybrid inheritance
  def showA(self):
    print('A')
class B(A):
  def showB(self):
    print("B")
class C(A):
  def showC(self):
    print('C')
class D(B,C):
  def showD(self):
    print('D')
obj=D()
obj.showD()
obj.showC()
obj.showB()
obj.showA()
```

#Polymorphism in Python

# Method Overriding

Method Overriding occurs when a child class provides its own implementation of a method that is already defined in the parent class, using the same method name and parameters.

```python
class Animal:
    def sound(self):   #Same method name sound() â†’ different behavior
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

# Polymorphism
a = [Dog(), Cat(), Animal()]

for i in a:
    i.sound()
```

```python
class Parent:
    def show(self):
        print("Parent class method")

class Child(Parent):
    def show(self):
        print("Child class method")

obj = Child()
obj.show()
```

```python
# Method Overriding
class A:
  def fun(self):
    print('data')
class B(A):
  def fun(self):
    print('science')
obj=B()
obj.fun()
```

#Method Overloading

Defining multiple methods with the same name but different parameters in the same class.

```python
class PPP:
  def fun(self,a,b=0,c=0):  #Same method name , Different number of arguments
    print(a+b+c)
obj=PPP()
obj.fun(10)
obj.fun(10,5)
obj.fun(10,5,15)
```

#Abstraction

Abstraction is the process of hiding internal implementation details and showing only necessary functionality.

```python
from abc import ABC,abstractmethod

class vehicle(ABC):

  @abstractmethod
  def start(self):
    pass

  @abstractmethod
  def stop(self):
    pass

class bike(vehicle):
  def start(self):
    print('bike started')

  def stop(self):
    print('bike stoped')
obj=bike()
obj.start()
obj.stop()
```

```python
from abc import ABC,abstractmethod
class WHATSUP1(ABC):  # abstract class
  def chat(self):
    print('chat is available')

  @abstractmethod
  def status(self):  # abstract function
    pass

class WHATSUP2(WHATSUP1):  # inherited
  def status(self):
    print('new change')
obj=WHATSUP2()
obj.status()
obj.chat()
```

```python
from abc import ABC, abstractmethod

class Bank(ABC):

    @abstractmethod
    def interest(self):
        pass

class SBI(Bank):
    def interest(self):
        print("SBI interest rate is 6%")

class HDFC(Bank):
    def interest(self):
        print("HDFC interest rate is 7%")

a = [SBI(), HDFC()]

for i in a:
    i.interest()
```
