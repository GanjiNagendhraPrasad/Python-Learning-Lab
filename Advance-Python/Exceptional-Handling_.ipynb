```python
import os
import sys


try:
  a=10
  b=0
  print(a/b)
except Exception as e:
  print(e)
  type,msg,line=sys.exc_info()
  print(type)
  print(msg)
  print(line.tb_lineno)   # here .tblineno to print error line no if not memory address will print
```
