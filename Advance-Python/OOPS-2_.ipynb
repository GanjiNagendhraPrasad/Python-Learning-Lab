**inheritance**

```python
# 1  single inheritance

class PARENT:
  def shop(self):
    print('i have shop')

class SON(PARENT):
  def job(self):
    print('i have job')

anil=SON()
anil.job()
anil.shop()
```

```python
# 2 multi level inheritance

class GRANDFATHER:
  def shop(self):
    print('i have shop')

class FATHER(GRANDFATHER):
  def land(self):
    print('i have land')

class SON(FATHER):
  def job(self):
    print('i have job')

anil=SON()
anil.job()
anil.land()
anil.shop()
```

```python
# 3 multiple inheritance

class FATHER:
  def shop(self):
    print('shop')

class MOTHER:
  def land(self):
    print('land')

class SON(FATHER,MOTHER):
  def job(self):
    print('job')

anil=SON()
anil.job()
anil.land()
anil.shop()
```

```python
# 4  Hierarichal inheritance

class FATHER:
  def shop(self):
    print('shop')

class B1(FATHER):
  def land(self):
    print('land')

class B2(FATHER):
  def job(self):
    print('job')

class B3(FATHER):
  def property(self):
    print('property')

anil=B3()
anil.property()
anil.shop()
```

```python
# 5 hybrid inheritance

class GRANDFATHER:
  def shop(self):
    print('shop')

class FATHER(GRANDFATHER):
  def land(self):
    print('land')

class MOTHER(GRANDFATHER):
  def property(self):
    print('property')

class SON(FATHER,MOTHER):
  def job(self):
    print('job')

anil=SON()
anil.job()
anil.property()
anil.land()
anil.shop()
```

```python
class PARENT():  # here () means connect to the object class
  def shop(self):
    print('i have shop')

class SON(PARENT):
  def job(self):
    print('i have job')

# object class take care of this
anil=SON()
anil.job()
anil.shop()
```

**polymorphisam**

```python
# 1 method overriding

def fun():
  print('java')

def fun():
  print('python')

fun()
```

```python
# 2 method overloading

def fun(a=None,b=None):
  if a==None and b==None:
    print('usa')
  elif a==None and b!=None:
    print('uk')
  elif a!=None and b==None:
    print('india')
  else:
    print('some where')

fun()
fun(a=10)
fun(b=10)
fun(10,20)
```

**Abstraction**

```python
from abc import ABC,abstractmethod

class WHATSAPPVERSION_11(ABC):  # abstract class, parent class
  def chat(self):
    print('chat is available')

  @abstractmethod
  def status(self):   # abstract method(function)
    pass

class WHATSAPPVERSION_22(WHATSAPPVERSION_11):  # inherit 1st version, child class
  def status(self):
    print('new changes')

obj=WHATSAPPVERSION_22()
obj.status()
```

```python
# example

from abc import ABC,abstractmethod

class VEHICLE(ABC):   # abstract class, parent class

  @abstractmethod
  def start(self):  # abstract function
    pass

  @abstractmethod
  def stop(self):  # abstract function
    pass

class BIKE(VEHICLE): # child class inherit parent class

  def start(self):
    print('bike started')

  def stop(self):
    print('bike stoped')

obj=BIKE()
obj.start()
obj.stop()
```
