```python
# advance level of reusability

class JJJ:
  a=10
  b=20

  def add(self):
    print(self.a + self.b)

  def sub(self):
    print(self.a - self.b)

  def common(self):
    self.add()
    self.sub()

anil=JJJ()
anil.common()
```

5. Simple Calculator using Reusability

```python
class PPP:
  a=10
  b=20

  def add(self):
    print(self.a+self.b)

  def sub(self):
    print(self.a-self.b)

  def mult(self):
    print(self.a*self.b)

  def div(self):
    print(self.a/self.b)


  def common(self):
    self.add()
    self.sub()
    self.mult()
    self.div()

anil=PPP()
anil.common()
```

```python
# without constructor find the prime numbers from 1 to 100

class PPP:

  def fun(self,a,b):

    for i in range(a,b):
      c=0
      for j in range(1,i+1):
        if i%j==0:
          c=c+1

      if c==2:
        print(f'prime {i}')
      else:
        print(f'not prime {i}')
anil=PPP()
anil.fun(1,101)
```

```python
# with constructor find the prime numbers from 1 to 100

class PPP:

  def __init__(self,a,b):
    self.a=a
    self.b=b

  def fun(self):
    for i in range(self.a,self.b):
      c=0
      for j in range(1,i+1):
        if i%j==0:
          c=c+1

      if c==2:
        print(f'prime {i}')

      else:
        print(f'not prime {i}')

anil=PPP(1,101)
anil.fun()
```

```python
import sys

class OOO:
  def __init__(self,a):
    try:
      self.a=a
    except Exception as e:
      ty,msg,lin=sys.exc_info()
      print(f'due to {ty} reason {msg} line no is {lin.tb_lineno}')

  def fun(self):
    try:
      c=0
      for i in  range(1,self.a+1):
        if self.a%i==0:
          c=c+1
      if c==2:
        print(f'prime {i}')
      else:
        print(f'not prime {i}')
    except Exception as e:
      ty,msg,lin=sys.exc_info()
      print(f'due to {ty} reason {msg} line no is {lin.tb_lineno}')


anil=OOO(7)
anil.fun()
```

```python
import sys

class PPP:

  def __init__(self,a,b):
    try:
      self.a=a
      self.b=b
    except Exception as e:
      msg,type,line=sys.exc_info()
      print(f'message : {msg} due to {type} line no is {line.tb_lineno}')
  def fun(self):
    try:
      for i in range(self.a,self.b): # (1,101)
        c = 0
        for j in range(1,i+1):
          if i % j == 0:
            c = c + 1
        if c == 2:
          print(f'Prime : {i}')
        else:
          print(f'Not Prime : {i}')
    except Exception as e:
      msg,type,line=sys.exc_info()
      print(f'message : {msg} due to {type} line no is {line.tb_lineno}')

anil=PPP(1,101)
anil.fun()
```

```python
# even no

import sys

class PPP:
  def __init__(self,a,b):
    try:
      self.a=a
      self.b=b
    except Exception as e:
      msg,type,line=sys.exc_info()
      print(f'message : {msg} due to {type} line no is {line.tb_lineno}')

  def fun(self):
    try:
      for i in range(self.a,self.b):
        if i%2==0:
          print(f'even {i}')
    except Exception as e:
      msg,type,line=sys.exc_info()
      print(f'message : {msg} due to {type} line no is {line.tb_lineno}')

anil=PPP(1,11)
anil.fun()
```
