```python
# functions in class

class LLL:
  f=10

  def fun():
    print('gm')

anil=LLL()
print(anil.f)  # calling only f
```

```python
class LLL:
  f=10

  def fun():   # here not accepting the copy of object
    print('gm')

anil=LLL()
print(anil.f)
anil.fun()
```

```python
class LLL:
  f=10

  def fun(b):  #  here  accepting the copy of object  with b variable
    print('gm')

anil=LLL()
print(anil.f)
anil.fun()
```

```python
class LLL:
  f=10

  def fun(b):
    print(b)
    print('gm')

anil=LLL()
print(anil)
print(anil.f)
anil.fun()
```

```python
class PPP:
  def fun(self):   # self keyword is used to hold copy of object
    print('gm')
anil=PPP()
anil.fun()
```

```python
class KKK:
  r=200

  def fun():
    print('gm')
    print(f'hii : {r}')

anil=KKK()
print(anil.r)  # calling r only
```

```python
class KKK:
  r=200

  def fun(self):
    print('gm')
    print(f'hii : {r}')

anil=KKK()
print(anil.r)
anil.fun()
```

```python
class KKK:
  r=200

  def fun(self):   # self is holding copy of object anil
    print('gm')
    print(f'hii : {self.r}')

anil=KKK()
print(anil.r)
anil.fun()
```

```python
# advance level of reusability

class JJJ:
  a=10
  b=20

  def add(self):
    print(self.a + self.b)

  def sub(self):
    print(self.a - self.b)

  def common(self):
    self.add()
    self.sub()

anil=JJJ()
anil.common()
```

1  Find Value of Ï€ (Pi) using Class Methods

  2. Square Root of a Number

3. Area of a Rectangle

```python
class LLL:
  length=10
  width=5

  def fun(self):
    print(f'area of rectangle is : {self.length*self.width}')

  def common(self):
    self.fun()

anil=LLL()
anil.common()
```

4. Area of a Circle

```python
class LLL:
  r=5

  def fun(self):
    print(f'area of circle is : {3.14 * self.r ** 2}')

  def common(self):
    self.fun()

anil=LLL()
anil.common()
```

5. Simple Calculator using Reusability

```python
class PPP:
  a=10
  b=20

  def add(self):
    print(self.a+self.b)

  def sub(self):
    print(self.a-self.b)

  def mult(self):
    print(self.a*self.b)

  def div(self):
    print(self.a/self.b)


  def common(self):
    self.add()
    self.sub()
    self.mult()
    self.div()

anil=PPP()
anil.common()
```

6. Find Square and Cube of a Number

```python
class LLL:
  n=5

  def sq(self):
    print(self.n**2)

  def cube(self):
    print(self.n**3)

  def common(self):
    self.sq()
    self.cube()

anil=LLL()
anil.common()
```

7. Perimeter of a Rectangle

```python
class LLL:
  length=10
  width=5

  def fun(self):
    print(2*(self.length+self.width))

  def mmm(self):
    self.fun()

anil=LLL()
anil.mmm()
```

8. Area of a Triangle

```python
class PPP:
  base=10
  height=15

  def area(self):
    print(f'area of triangle is : {0.5*(self.base*self.height)}')

  def common(self):
    self.area()

anil=PPP()
anil.common()
```

9. Convert Celsius to Fahrenheit

```python
class PPP:
  c=30

  def fun(self):
    print((self.c*9/5)+32)

  def common(self):
    self.fun()

anil=PPP()
anil.common()
```

10. Find Average of Numbers

```python
class PPP:
  c=[1,2,3,4]

  def fun(self):
    s=sum(self.c)
    c=len(self.c)
    print(s/c)


  def common(self):
    self.fun()

anil=PPP()
anil.common()
```
