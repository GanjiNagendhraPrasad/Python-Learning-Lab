```python
class CAL:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def add(self):
    print(self.a+self.b)
  def sub(self):
    print(self.a-self.b)
  def mul(self):
    print(self.a*self.b)
  def common(self):
    self.add()
    self.sub()
    self.mul()
obj=CAL(10,20)
obj.common()
```

```python
# without constructor
class PRIME:
  def fun(self,a,b):
    for i in range(a,b):
      c=0
      for j in range(1,i+1):
        if i%j==0:
          c=c+1
      if c==2:
        print('prime',i)
      else:
        print('not prime',i)
obj=PRIME()
obj.fun(1,21)
```

```python
# even
import sys
class PPP:
  def __init__(self,a,b):
    try:
      self.a=a
      self.b=b
    except Exception as e:
      typ,msg,lin=sys.exc_info()
      print(f'type is {typ} due to {msg} line no is {lin.tb_lineno}')
  def fun(self):
    try:
      for i in range(self.a,self.b):
        if i%2==0:
          print(f'even {i}')
    except Exception as e:
      typ,msg,lin=sys.exc_info()
      print(f'type is {typ} due to {msg} line no is {lin.tb_lineno}')
obj=PPP(1,11)
obj.fun()
```

10 Find Average of Numbers

```python
import sys
class PPP:
  def __init__(self,a):
    try:
      self.a=a
    except Exception as e:
      typ,msg,lin=sys.exc_info()
      print(f'type is {typ} due to {msg} line no is {lin.tb_lineno}')
  def fun(self):
    try:
      x=sum(self.a)
      y=len(self.a)
      print('avg is : ',x/y)
    except Exception as e:
      typ,msg,lin=sys.exc_info()
      print(f'type is {typ} due to {msg} line no is {lin.tb_lineno}')
obj=PPP([1,2,3,4])
obj.fun()
```
