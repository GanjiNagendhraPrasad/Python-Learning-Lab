**TYPE CONERSION**

```python
# int to float
a=10
print(a)
print(type(a))

b=float(a)
print(b)
print(type(b))
```

```python
# int to string
a=10
print(a)
print(type(a))

b=str(a)
print(b)
print(type(b))
```

```python
# float to int
a=11.5
print(a)
print(type(a))

b=int(a)
print(b)
print(type(b))
```

```python
# float to string
a=11.5
print(a)
print(type(a))

b=str(a)
print(b)
print(type(b))
```

```python
# string to int
a='ppp'
print(a)
print(type(a))

b=int(a)
print(b)
print(type(b))
```

```python
# string to float
a='pp123'
print(a)
print(type(a))

b=float(a)
print(b)
print(type(b))
```
