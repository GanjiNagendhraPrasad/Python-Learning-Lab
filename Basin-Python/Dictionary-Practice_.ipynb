```python
a={'gas':79,'petrol':89,'disel':99}
b=a  # deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.update({'gas':999})

print(a)
print(b)
```

```python
a={'gas':79,'petrol':89,'disel':99}
b=a.copy()  # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.update({'gas':999})

print(a)
print(b)
```
