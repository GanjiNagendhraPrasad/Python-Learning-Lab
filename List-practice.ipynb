```python
a=[22,12,43,32,19]           # -COPY-
b=a                        # DEEP COPY
print(a)
print(b)                 # here same ADDRESS

print(id(a))
print(id(b))

a.append(50)

print(a)
print(b)
```

```python
a=[22,12,43,32,19]     # SHALLOW COPY
b=a.copy()

print(a)              # here ADDRESS is NOT same
print(b)

print(id(a))
print(id(b))

a.append(50)

print(a)
print(b)
```
