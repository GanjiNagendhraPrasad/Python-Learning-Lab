```python
def even(x):
  if x%2==0:
    return True
  else:
    return False
def div(y):
  if even(y) and y%3==0:
    return True
  else:
    return False

print(div(6))
```

```python
def fun(x):    # factorial
  if x==0 or x==1:
    return 1
  else:
    return x*fun(x-1)
print(fun(5))
```
