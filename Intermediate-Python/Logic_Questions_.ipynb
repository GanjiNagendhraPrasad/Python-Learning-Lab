# reverse

```python
n=123
a=n%10 # last no
b=n//10 # remaining 2 no's

c=b%10
d=b//10

x=str(a)+str(c)+str(d)
print(x)
```

# palindrome

```python
n=121
a=n%10 # last no
b=n//10 # remaining 2 no's

c=b%10
d=b//10

x=str(a)+str(c)+str(d)
z=int(x)
if n==z:
  print('palindrome',n)
else:
  print('not palindrome',n)
```

```python
f=121
t=f
s=0
while f>0:
  r=f%10
  s=(s*10)+r
  f=f//10
if s==t:
  print('palindrome',t)
else:
  print('not palindrome',t)
```

#perfect

#Amstrong

```python
f=153
t=f
s=0
while f>0:
  r=f%10
  s=s+(r*r*r)
  f=f//10
if s==t:
  print('Amstrong',t)
else:
  print('not Amstrong',t)
```

#strong number

```python
import math
f=145
t=f
s=0
while f>0:
  r=f%10
  s=s+math.factorial(r)
  f=f//10
if s==t:
  print('strong',t)
else:
  print('not strong',t)
```

#fibonacci series

```python
n=8
a=0
b=1
print(a)
print(b)
for i in range(n-2):
  c=a+b
  print(c)
  a=b
  b=c
```

#prime number

```python
n=7
c=0
for i in range(1,n+1):
    if n%i==0:
        c=c+1
if c==2:
    print('prime',n)
else:
    print('not prime',n)
```
