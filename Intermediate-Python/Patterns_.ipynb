```python
for i in range(1,5):
  for j in range(1,6):
    print('*',end=' ')
  print()
```

#Right Triangle

```python
for i in range(1,5):
  for j in range(1,i+1):
    print('*',end=' ')
  print()
```

# Reverse Right Triangle

```python
for i in range(4,0,-1):
  for j in range(1,i+1):
    print('*',end=' ')
  print()
```

#Left Triangle

```python
for i in range(1,5):
  print(' '*(4-i) + '*'*i)
```

# reverse left triangle

```python
for i in range(4,0,-1):
  print(' '*(4-i)  +  '*'*i)
```

#Pyramid Pattern

```python
for i in range(1,6):
  print(' '*(5-i)   +   '*'*(2*i-1))
```

#Inverted Pyramid

```python
for i in range(6,0,-1):
  print(' '*(5-i)   +   '*'*(2*i-1))
```

#Diamond Pattern

```python
for i in range(1,6): # n-1 logic
  print(' '*(5-i)   +   '*'*(2*i-1))
for i in range(5,0,-1):
  print(' '*(5-i)   +   '*'*(2*i-1))
```
