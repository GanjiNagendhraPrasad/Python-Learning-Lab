***Modularization using functions***

**Arguments**

-   * args  # single star
- ** keyword args [kwargs]  # double star

```python
def kiran(a,b):
  print(a)
  print(b)
kiran([10,20,30],[1,2,3])
```

```python
def kiran(*a):  # 1 variable using
  print(a)
kiran([10,20,30],[1,2,3])
```

```python
# ** args

def sharuk(**a):
  print(a)
sharuk(apple=100,banaba=200,carrot=300)
```

**Lambda functions**

```python
f = lambda a,b : a + b
print(f(10,20))
```

```python
n=2
f=lambda x,y : x*y
for i in range(1,11):
  print(f'{n}*{i}={f(n,i)}')
```

map

```python
def fun(b):
  if b%2==0:
    return b
  else:
    return 'usa'
sol=map(fun,[10,11,12,13,14,15])
print(list(sol))
```

```python
sol=map(lambda x:x%2==0 , [10,11,12,13,14,15])
print(list(sol))
```

filter

```python
def fun(b):
  if b%2==0:
    return b

sol=filter(fun,[10,11,12,13,14,15])
print(list(sol))
```

reduce

```python
import functools

f=[1,4,7,9,3,6]

sol=functools.reduce(lambda x,y : x+y , [1,4,7,9,3,6])
print(sol)
```

map with filter

```python
sol=map(lambda x:x+20 , filter(lambda x:x%2==0 , [10,11,21,22,26,36,35]))
print(list(sol))
```

filter with map

```python
sol=filter(lambda x:x%2==0 , map(lambda x:x+20 , [10,11,21,22,26,36,35]))
print(list(sol))
```

reduce with map & filter

```python
import functools

s=functools.reduce(lambda x,y : x+y , map(lambda x:x+20 , filter(lambda x:x%2==0 , [10,11,21,22,26,36,35])))
print(s)
```

**COMPREHENSIONS**

List comprehensions

```python
print(['gm'  for i in range(1,6)])
```

```python
print([f'even:{i}' for i in range(11,16) if i%2==0])
```

```python
print([f'even {i}' if i%2==0 else f'odd"{i}' for i in range(11,16)])
```

dictionary comprehension

```python
f={'apple':100,'banana':200,'carrot':300}

for i in f:
  f[i]=f[i]*2
print(f)
```

```python
f={'apple':100,'banana':200,'carrot':300}
print({i:f[i]*2 for i in f})
```

```python
f={'apple':100,'banana':200,'carrot':300}

d={}

for i in f:
  if f[i]>150:
    d[i]=f[i]*3
print(d)
```

```python
f={'apple':100,'banana':200,'carrot':300}
print({i:f[i]*3 for i in f if f[i]>150})
```

```python
f={'apple':100,'banana':200,'carrot':300}
for i in f:
  if f[i]>150:
    f[i]=f[i]*3
  else:
    f[i]=f[i]*2
print(f)
```

```python
print({i:(j*3 if j>150 else j*2) for i,j in f.items()})
```

Calculate Time

```python
import time
st=time.time()
f={'apple':100,'banana':200,'carrot':300}
for i in f:
  if f[i] > 150:
    f[i] = f[i]*3
  else:
    f[i] = f[i]*2
print(f)
print(f'tt : {time.time() - st}')
```

```python
import time
st=time.time()
print({i:(j*3 if j>150 else j*2) for i,j in f.items()})
print(f'tt : {time.time() - st}')
```
