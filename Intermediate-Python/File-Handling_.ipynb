**File handling**

```python
a=open('prasad.txt','w')  # createing a new filr
a.write('data science')  # adding data
a.close()
```

```python
a=open('data.txt','w')  # emplty file
a.write('java java ')  # add data
a.close()
```

```python
a=open('naga.txt','w')  # file exist with data
a.write('python python')  # new data override
a.close()
```

mode **a**

```python
a=open('prasad.txt','a')
a.write('data science11')  # data will append to the existing file with existing data
a.close()
```

WITH

```python
with open ('naga.txt','w') as a:
  a.write('data')   # here no need of close function
```

```python
with open('data.txt','a') as a:
  a.write('aiml')
```

dump

```python
import json
a={'apple':10,'bat':20,'cat':30}
with open ('kumar.json','w') as x:
  json.dump(a,x)   # dict to json
```

load

```python
with open('kumar.json','r') as a:
  r=json.load(a)    # read dict from json file
print(r)
```

dumps

```python
a={'apple':10,'bat':20,'cat':30}
print(a)
print(type(a))

r=json.dumps(a)  # dict to str
print(r)
print(type(r))
```

loads

```python
a={'apple':10,'bat':20,'cat':30}
print(a)
print(type(a))

r=json.dumps(a)  # dict to str
print(r)
print(type(r))

x=json.loads(r)  # str to dict
print(x)
print(type(x))
```
