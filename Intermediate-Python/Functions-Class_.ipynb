```python
def fun(n):
  s=0
  for i in range(1,n):
    if n%i==0:
      s=s+i
  if s==n:
    print(f'perfect : {n}')
  else:
    print(f'not perfect : {n}')

fun(6)
```

```python
# global variable

a=100 # out of function

def fun():
  print(a)
fun()

print(a)
```

```python
# local variable

def fun():
  r=100   # local variable
  print(r)
fun()

print(r)   # here error bcz r is inside function
```

```python
def fun():
  global p   # global means converting local to global variable then we can use anywhere
  p=900
  print(p)
fun()

print(p)
```

```python
def fun(n):
  for i in range(1,7):
    if i % 2==0:
      print(i)

  c=0
  for i in range(1,n+1):
    if n%i==0:
      c=c+1
  if c==2:
    print('prime')
  else:
    print('not prime')

fun(7)
```

```python
def fun():
  for i in range(1,10):
    if i % 2 == 0:
      return i  # here return will execute 1 time only then stops.
a=fun()
print(a)
```

```python
def fun():
  a=[]
  for i in range(1,10):
    if i%2==0:
      a.append(i)
  return a

a=fun()
print(a)
```

```python
def prasad(a,b):
  n=[]    # here variables be different , give different names
  for i in range(a,b):
    if i%2==0:
      n.append(i)
  return n
a=prasad(1,11)
print(a)
```
