```python
n=6
t=0
for i in range(1,n):
  if n%i==0:
    t=t+i
print(f'{n} is a perfect number :',t==n)
```

```python
n=6
t=[]
for i in range(1,n):
  if n%i==0:
    t.append(i)

if sum(t)==n:
  print(f'{n} is a perfect num')
else:
  print(f'{n} is not a perfect num')
```

PALINDROME

```python
f=121
t=f
s=0
while f>0:
  r=f%10   # TO GET LAST VALUE
  s=(s*10)+r
  f=f//10   # TO GET REMAINING VALUES
if t==s:
  print('palindrome')
else:
  print('not palindrome')
```

AMSTRONG

```python
f=153
t=f
s=0
while f>0:
  r=f%10
  s=s+(r*r*r)
  f=f//10
if t==s:
  print('amstrong')
else:
  print('not amstrong')
```

STRONG NUMBER OR NOT

```python
import math
f=145
t=f
s=0
while f>0:
  r=f%10
  s=s+math.factorial(r)
  f=f//10
if t==s:
  print('strong')
else:
  print('not strong')
```

fibonacci series

```python
n=8
a=0
b=1
print(a)
print(b)
for i in range(n-2):
  c=a+b  # 1
  print(c)
  a=b
  b=c
```
